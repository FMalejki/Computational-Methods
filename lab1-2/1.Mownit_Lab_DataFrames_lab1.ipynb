````julia
using LinearAlgebra, DataFrames, CSV, Random, Plots, Statistics

function measure_time(f, args...; repeats=10)
    times = Float64[]
    for _ in 1:repeats
        t = @elapsed f(args...)
        push!(times, t)
    end
    return times
end
````

````julia
sizes = [500, 1000, 2000, 3000, 4000]

results = DataFrame(n=Int[], operation=String[], time=Float64[])

for n in sizes
    Random.seed!(42)
    A = rand(n, n)
    x = rand(n)
    y = rand(n)
    
    dot_times = measure_time((x, A, y) -> dot(x, A * y), x, A, y)
    append!(results, DataFrame(n=fill(n, 10), operation=fill("dot(x, A, y)", 10), time=dot_times))
    
    matvec_times = measure_time((A, x) -> A * x, A, x)
    append!(results, DataFrame(n=fill(n, 10), operation=fill("A*x", 10), time=matvec_times))
end
````

````julia
CSV.write("benchmark_results.csv", results)

df = CSV.read("benchmark_results.csv", DataFrame)
````

````julia
grouped = combine(groupby(df, [:n, :operation]),
                  :time => mean => :mean_time,
                  :time => std => :std_time)

plot1 = plot(grouped[grouped.operation .== "dot(x, A, y)", :n], 
             grouped[grouped.operation .== "dot(x, A, y)", :mean_time],
             yerr=grouped[grouped.operation .== "dot(x, A, y)", :std_time],
             label="dot(x, A, y)", xlabel="Rozmiar macierzy n", ylabel="Czas [s]",
             title="Czas obliczeń dla dot(x, A, y)", markershape=:circle)

plot2 = plot(grouped[grouped.operation .== "A*x", :n], 
             grouped[grouped.operation .== "A*x", :mean_time],
             yerr=grouped[grouped.operation .== "A*x", :std_time],
             label="A*x", xlabel="Rozmiar macierzy n", ylabel="Czas [s]",
             title="Czas obliczeń dla A*x", markershape=:circle)
````

````julia
plot(plot1, plot2, layout=(1, 2), size=(900, 400))
````
