### Tablice wielowymiarowe i mnożenie macierzy w Julii

```julia
# tablice wielowymiarowe w Julii
Asmall = [[1.0 4.0]; [0.0 1.0]]
Bsmall = Asmall
# size(B,1)
```

#### Wersja naiwna

```julia
# mnożenie macierzy - wersja naiwna
function naive_multiplication(A,B)
    C = zeros(Float64, size(A,1), size(B,2))
    for i=1:size(A,1)
        for j=1:size(B,2)
            for k=1:size(A,2)
                C[i,j] = C[i,j] + A[i,k]*B[k,j]
            end
        end
    end
    C
end
```

```julia
#kompilacja
naive_multiplication(Asmall, Bsmall)
```

```julia
#kompilacja
@elapsed naive_multiplication(Asmall, Bsmall)
```

```julia
# poprawiona funkcja korzytająca z powyższego oraz z faktu, że
# można zmieniać kolejność operacji dodawania (a co za tym idzie kolejnosc petli).
function better_multiplication(A, B)
    C = zeros(Float64, size(A,1), size(B,2))
    for j=1:size(B,2)
        for k=1:size(A,2)
            for i=1:size(A,1)
                C[i,j] = C[i,j] + A[i,k]*B[k,j]
            end
        end
    end
    C
end
```

```julia
@elapsed better_multiplication(Asmall, Bsmall)
```

```julia
naive_res = Float64[]
better_res = Float64[]
blas_res = Float64[]

i = 100
while i <= 1000
    A = rand(i, i)
    B = rand(i, i)
    println("Generated matrix A of size $i x $i")
    println("Generated matrix B of size $i x $i")

    push!(naive_res, @elapsed naive_multiplication(A, B))
    push!(better_res, @elapsed better_multiplication(A, B))
    push!(blas_res, @elapsed A * B)

    i += 100
end
```

```julia
#Pkg.add("Statistics")
import Statistics
using Statistics
function standard_error(data)
    n = length(data)
    return std(data) / sqrt(n)
end
```

```julia
naive_se = standard_error(naive_res)
better_se = standard_error(better_res)
blas_se = standard_error(blas_res)
```

```julia
plot(100:100:1000, naive_res,yerror=fill(naive_se, length(100:100:1000)), label = "naive multiplication", xlabel="Matrix Size", ylabel="Time (s)")
plot!(100:100:1000,better_res,yerror=fill(naive_se, length(100:100:1000)), label = "better multiplication")
plot!(100:100:1000,blas_res,yerror=fill(naive_se, length(100:100:1000)), label = "built in BLAS multiplication")
```

```julia
#include <stdio.h>
#include <stdlib.h>
#include <time.h>
#include <gsl/gsl_blas.h>

void generate_random_matrix(double *matrix, int n) {
    for (int i = 0; i < n * n; i++) {
        matrix[i] = (double)rand() / RAND_MAX;
    }
}

void naive_multiplication(double *A, double *B, double *C, int n) {
    for (int i = 0; i < n; i++) {
        for (int j = 0; j < n; j++) {
            C[i * n + j] = 0.0;
            for (int k = 0; k < n; k++) {
                C[i * n + j] += A[i * n + k] * B[k * n + j];
            }
        }
    }
}

void better_multiplication(double *A, double *B, double *C, int n) {
    for (int i = 0; i < n * n; i++) {
        C[i] = 0.0;
    }
    
    for (int j = 0; j < n; j++) {
        for (int k = 0; k < n; k++) {
            for (int i = 0; i < n; i++) {
                C[i * n + j] += A[i * n + k] * B[k * n + j];
            }
        }
    }
}

void blas_multiplication(double *A, double *B, double *C, int n) {
    gsl_matrix_view A_view = gsl_matrix_view_array(A, n, n);
    gsl_matrix_view B_view = gsl_matrix_view_array(B, n, n);
    gsl_matrix_view C_view = gsl_matrix_view_array(C, n, n);
    
    // C = alpha*A*B + beta*C, gdzie alpha=1.0, beta=0.0
    gsl_blas_dgemm(CblasNoTrans, CblasNoTrans, 1.0, &A_view.matrix, &B_view.matrix, 0.0, &C_view.matrix);
}

double measure_time(void (*func)(double*, double*, double*, int), double *A, double *B, double *C, int n) {
    clock_t start, end;
    start = clock();
    func(A, B, C, n);
    end = clock();
    return ((double)(end - start)) / CLOCKS_PER_SEC;
}

int main() {
    srand(time(NULL));
    
    int sizes[] = {100, 200, 300, 400, 500, 600, 700, 800, 900, 1000};
    int num_sizes = sizeof(sizes) / sizeof(sizes[0]);
    
    printf("Rozmiar,Naiwna,Ulepszona,BLAS\n");
    
    for (int s = 0; s < num_sizes; s++) {
        int n = sizes[s];
        
        double *A = (double*)malloc(n * n * sizeof(double));
        double *B = (double*)malloc(n * n * sizeof(double));
        double *C = (double*)malloc(n * n * sizeof(double));
        
        generate_random_matrix(A, n);
        generate_random_matrix(B, n);
        
        //Pomiary
        double time_naive = measure_time(naive_multiplication, A, B, C, n);
        
        double time_better = measure_time(better_multiplication, A, B, C, n);
        
        double time_blas = measure_time(blas_multiplication, A, B, C, n);
        
        printf("%d,%f,%f,%f\n", n, time_naive, time_better, time_blas);
        
        // Zwolnienie pamięci
        free(A);
        free(B);
        free(C);
    }
    
    return 0;
}
```
